```python
from google.colab import drive
import pandas as pd

# Mount Google Drive
drive.mount('/content/drive')

diabetes_data = pd.read_csv('/content/drive/My Drive/diabetes.csv')
diabetes_data
```

```python
from sklearn.neighbors import KNeighborsClassifier
data = list(zip(diabetes_data['Pregnancies'],diabetes_data['Glucose'],diabetes_data['BloodPressure'],diabetes_data['SkinThickness']))
print(data)
```

```python
y = diabetes_data.Outcome
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(data,y,test_size=0.2,random_state=10)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train,y_train)
y_predict = model.predict(X_test)
y_predict_train=model.predict(X_train)
accuracy_score(y_test,y_predict)
```

```python
accuracy_score(y_train,y_predict_train)
```

```python
print( accuracy_score(y_test,y_predict))
```

```python
print(classification_report(y_test,y_predict))
```

```python
confusion_matrix(y_test,y_predict)
```
