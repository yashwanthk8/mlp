dropna(): Removes rows or columns with missing values.

fillna(): Fills missing values with a specified value.

```
data={'A':[1,2,None,4],'B':[5,None,7,8],'C':[None,10,11,12],'D':[13,14,15,16]}
df=pd.DataFrame(data)
print(df)
print(df['B'])
df_dropped=df.dropna()
print(df_dropped)
df_filled=df.fillna(0)
print(df_filled)
```

```
print(df)
df.fillna(99,inplace=True)
print(df)
```

replace by mean median and calculating mean ,median



```
data={'A':[1,2,None,4],'B':[5,None,7,8],'C':[None,10,11,12],'D':[13,14,15,None]}
df1=pd.DataFrame(data)
print(df1)
df1['C'].fillna(66,inplace=True)
x=df1['B'].mean()
df1['B'].fillna(x,inplace=True)
y=df1['A'].median()
df1['A'].fillna(y,inplace=True)
print(df1)
z=df1['D'].mode()[0]
df1['D'].fillna(z,inplace=True)
print(df1)
```
